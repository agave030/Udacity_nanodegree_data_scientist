Here we choose first way to observe and remove outlier.

```python
# We're using the price column and there's no NaN value.
# No need to data cleaning.
# Now we're going to remove outlier by calculating mean +/- 3 std.
# I define an empty list(delete) for appending the outlier value, so we can drop it.

delete = []
for i in range(len(listing['price'])):
    if listing['price'].iloc[i] > (listing['price'].mean() + 3 * listing['price'].std()) or listing['price'].iloc[i] < (listing['price'].mean() - 3 * listing['price'].std()):
        delete.append(listing['price'].index[i])
listing = listing.drop(delete)
```

Then, we're going to calculate the distance from airbnb to airport. First, we define a function for converting the latitude and longitude to absolute distance.

```python
def getDistance(latA, lonA, latB, lonB):
    ra = 6378140  # radius of equator: meter
    rb = 6356755  # radius of polar: meter
    flatten = (ra - rb) / ra  # Partial rate of the earth
    # change angle to radians
    radLatA = math.radians(latA)
    radLonA = math.radians(lonA)
    radLatB = math.radians(latB)
    radLonB = math.radians(lonB)

    pA = math.atan(rb / ra * math.tan(radLatA))
    pB = math.atan(rb / ra * math.tan(radLatB))
    x = math.acos(math.sin(pA) * math.sin(pB) + math.cos(pA) * math.cos(pB) * math.cos(radLonA - radLonB))
    c1 = (math.sin(x) - x) * (math.sin(pA) + math.sin(pB)) ** 2 / math.cos(x / 2) ** 2
    c2 = (math.sin(x) + x) * (math.sin(pA) - math.sin(pB)) ** 2 / math.sin(x / 2) ** 2
    dr = flatten / 8 * (c1 - c2)
    distance = ra * (x + dr)
    distance = round(distance / 1000, 6)
    return distance
```

d is distance between two poing. f = flatten. latA = Phi1, lonA = lambda1, latB =phi2, lonB = lambda2. The latitude and longitude need to be converted into radian.

```python
for index in range(len(listing)):
    distance.append(getDistance(tacoma_airport[0], tacoma_airport[1], listing['latitude'].iloc[index], listing['longitude'].iloc[index]))
    
```
